```python
#Make copy of dataframe, dropping Subject Arrest Charges
new_df = df.copy().drop(columns="Subject Arrest Charges")
```

```python
#Fill null values with "Unknown"
new_df.fillna("Unknown", inplace=True)
```

```python
new_df.replace({'Unknown Working Status':'Unknown'}, inplace=True)
```

```python
#Cast Number of Officers and Number of Subjects columns as string type
new_df = new_df.astype({'Number of Officers':str, 'Number of Subjects':str})
```

```python
#Split strings in each column with "|" delimiter
#Transformed columns contain list of values
for col in new_df.columns:
    if any(new_df[col].str.contains("|")):
        new_df[f"{col}_transformed"] = new_df[col].str.split("|")
```

```python
#Drop original multi-value columns
new_df.drop(columns=['Use of Force Type', 'Use of Force Level', 'Use of Force Effective', 'Officer Race/Ethnicity', 'Officer Gender', 'Officer Age',
                    'Officer Years of Service', 'Distance Between', 'Subject Gender', 'Subject Ethnicity', 'Subject Age', 'Subject Build', 
                    'Subject Influencing Factors', 'Subject Height', 'Subject Injured', 'Subject Hospitalized', 'Subject Arrested', 'Officer Injured'], inplace=True)
```

```python
#Explode officer dataframe
officer_df = officer_df.explode(column=['Use of Force Type_transformed',
       'Use of Force Level_transformed', 'Use of Force Effective_transformed',
       'Officer Race/Ethnicity_transformed', 'Officer Gender_transformed',
       'Officer Age_transformed', 'Officer Years of Service_transformed',
       'Officer Injured_transformed'])
```

```python
#Remove "_transformed" suffix from columns
officer_df.rename(columns=lambda x: x.replace('_transformed', ''), inplace=True)
```

```python
max_lengths = subject_df.apply(lambda row: max(len(row[col]) if isinstance(row[col], list) else 0 for col in subject_df.columns), axis=1)
```

```python
#Define function to pad with "Unknown" to match the max length of any value per row
def pad_with_unknowns(row, max_length):
    padded_row = {}
    for col in subject_df.columns:
        values = row[col]
        if isinstance(values, list):
            #Pad the list if it's shorter than the maximum length
            padded_row[col] = values + ['Unknown'] * (max_length - len(values))
        else:
            #Handle cases where the value is not a list (will also replace ID column)
            padded_row[col] = ['Unknown'] * max_length
    return padded_row
```

```python
#Apply function to standardize rows in subject_df
standardized_rows = subject_df.apply(lambda row: pad_with_unknowns(row, max_lengths[row.name]), axis=1)
standardized_df = pd.DataFrame(list(standardized_rows))
```

```python
#Fix PIB File Number since the function replaced them all
standardized_df['PIB File Number'] = subject_df['PIB File Number']
```

```python
subject_df = standardized_df
```
